```python
#importing required libraries
import numpy as np
import scipy as sp
import pandas as pd
```

```python
#Question 1
#Add column called ticker(META)
df1 = pd.read_csv("Documents/BU/DSC 350/Week 5/facebook.csv")
df1.insert(0, column = "Ticker", value = "META")
df1.head()
```

```python
#Add column called ticker(AMZN)
df2 = pd.read_csv("Documents/BU/DSC 350/Week 5/amazon.csv")
df2.insert(0, column = "Ticker", value = "AMZN")
df2.head()
```

```python
#Add column called ticker(AAPL)
df3 = pd.read_csv("Documents/BU/DSC 350/Week 5/apple.csv")
df3.insert(0, column = "Ticker", value = "AAPL")
df3.head()
```

```python
#Add column called ticker(NFLX)
df4 = pd.read_csv("Documents/BU/DSC 350/Week 5/netflix.csv")
df4.insert(0, column = "Ticker", value = "NFLX")
df4.head()
```

```python
#Add column called ticker(GOOG)
df5 = pd.read_csv("Documents/BU/DSC 350/Week 5/google.csv")
df5.insert(0, column = "Ticker", value = "GOOG")
df5.head()
```

```python
#Append them together into a single dataframe
dfs = [df1, df2, df3, df4, df5]
faang = pd.concat(dfs)
faang.head()
```

```python
faang.tail()
```

```python
#Question 2.
#Use type conversion to change the date column into datetime
faang.loc[:,'date'] = pd.to_datetime(faang.date)
faang.head()
```

```python
#Use type conversion to change the volume column into integers
faang.loc[:,'volume'] = pd.to_numeric(faang.volume)
faang.head()
```

```python
#Sort by date and ticker columns
faang.sort_values(by=['date', 'Ticker'], ascending=True).head(10)
```

```python
#Question 3
#Highest 7 rows for volume
faang.nlargest(n=7, columns='volume')
```

```python
#Question 4
#Use melt to change to long format
melted_faang = faang.melt(id_vars=['date', 'Ticker'], value_vars=['high', 'low', 'open', 'close',],
                         value_name='price', var_name='rate')
melted_faang.head()
```
